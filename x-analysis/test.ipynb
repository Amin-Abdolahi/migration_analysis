```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('sample_posts.csv')

# Count posts per username
print("Number of posts per username:")
print(df['username'].value_counts())

# Filter posts with likes > 200
print("\nPosts with more than 200 likes:")
print(df[df['likes'] > 200][['text', 'likes']])
```

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('sample_posts.csv')

# Calculate average likes
print("Average number of likes:")
print(df['likes'].mean())

# Sort posts by likes (descending)
print("\nPosts sorted by likes (highest to lowest):")
print(df.sort_values(by='likes', ascending=False)[['text', 'likes']])
```

```python
# Import libraries
import pandas as pd
from textblob import TextBlob

# Load the CSV file
df = pd.read_csv('sample_posts.csv')

# Function to calculate sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity  # Returns a value between -1 (negative) and 1 (positive)

# Apply sentiment analysis to the 'text' column
df['sentiment'] = df['text'].apply(get_sentiment)

# Display text, likes, and sentiment
print("Posts with sentiment scores:")
print(df[['text', 'likes', 'sentiment']])
```

```python
# Categorize sentiment
def sentiment_category(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Add sentiment category column
df['sentiment_category'] = df['sentiment'].apply(sentiment_category)

# Display text, likes, sentiment, and category
print("\nPosts with sentiment categories:")
print(df[['text', 'likes', 'sentiment', 'sentiment_category']])

# Count posts in each category
print("\nCount of posts by sentiment category:")
print(df['sentiment_category'].value_counts())
```

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load the CSV file
df = pd.read_csv('sample_posts.csv')

# Calculate sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis
df['sentiment'] = df['text'].apply(get_sentiment)

# Categorize sentiment
def sentiment_category(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_category'] = df['sentiment'].apply(sentiment_category)

# Plot count of posts by sentiment category
plt.figure(figsize=(8, 5))
sns.countplot(x='sentiment_category', data=df, order=['Positive', 'Neutral', 'Negative'])
plt.title('Count of Posts by Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Posts')
plt.show()
```

```python
# Import libraries
import pandas as pd
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('sample_posts.csv')

# Calculate sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis
df['sentiment'] = df['text'].apply(get_sentiment)

# Categorize sentiment
def sentiment_category(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_category'] = df['sentiment'].apply(sentiment_category)
```

```python
# Calculate average likes per sentiment category
print("Average likes by sentiment category:")
print(df.groupby('sentiment_category')['likes'].mean())

# Boxplot to show distribution of likes by sentiment category
plt.figure(figsize=(8, 5))
sns.boxplot(x='sentiment_category', y='likes', data=df, order=['Positive', 'Neutral', 'Negative'])
plt.title('Distribution of Likes by Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Likes')
plt.show()
```

```python
# Import libraries
import pandas as pd
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('sample_posts.csv')

# Calculate sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis
df['sentiment'] = df['text'].apply(get_sentiment)

# Categorize sentiment
def sentiment_category(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_category'] = df['sentiment'].apply(sentiment_category)

# Calculate average likes per sentiment category
print("Average likes by sentiment category:")
average_likes = df.groupby('sentiment_category')['likes'].mean()
print(average_likes)
```

```python
# Bar plot for average likes by sentiment category
plt.figure(figsize=(8, 5))
sns.barplot(x=average_likes.index, y=average_likes.values, order=['Positive', 'Neutral', 'Negative'])
plt.title('Average Likes by Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Average Number of Likes')
plt.show()
```
