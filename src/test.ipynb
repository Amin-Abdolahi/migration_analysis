```python
import pandas as pd
import glob
import os

# مسیر پوشه‌ای که فایل‌های CSV داخلش هست
folder_path = "../data/raw"

# پیدا کردن همه فایل‌های CSV در پوشه
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

# لیست برای ذخیره دیتافریم‌ها
dfs = []
dfs = []

for file in all_files:
    try:
        df = pd.read_csv(file, sep=";")  # جداکننده ; چون داده‌هات اینجوری بود
        dfs.append(df)
    except Exception as e:
        print(f"مشکل در خواندن {file}: {e}")

# ترکیب همه دیتافریم‌ها
combined_df = pd.concat(dfs, ignore_index=True)

# ذخیره به عنوان یک فایل CSV جدید
output_path = os.path.join(folder_path, "combined_data.csv")
combined_df.to_csv(output_path, index=False)

print(f"فایل ترکیب‌شده ذخیره شد: {output_path}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('../data/raw/combined_data.csv')

# Filter rows for foreign nationals (Ausland und Sonstige) and Bundesländer
foreign_data = df[(df['2_variable_attribute_label'] == 'Ausland und Sonstige') & 
                (df['1_variable_label'] == 'Bundesländer')]

# Separate percentage and count data
foreign_percent = foreign_data[foreign_data['value_unit'] == '%'][['1_variable_attribute_label', 'value']]
foreign_count = foreign_data[foreign_data['value_unit'] == 'Anzahl'][['1_variable_attribute_label', 'value']]

# Replace comma with dot in percentage values and convert to numeric
foreign_percent['value'] = foreign_percent['value'].str.replace(',', '.').astype(float)
foreign_count['value'] = pd.to_numeric(foreign_count['value'], errors='coerce')

# Rename columns
foreign_percent.columns = ['State', 'Percentage']
foreign_count.columns = ['State', 'Count']

# Remove duplicates by taking the latest entry
foreign_percent = foreign_percent.groupby('State').last().reset_index()
foreign_count = foreign_count.groupby('State').last().reset_index()

# Display filtered data
print("Percentage of foreign nationals by state:")
print(foreign_percent)
print("\nNumber of foreign nationals by state:")
print(foreign_count)

# Set font to avoid font errors (use system default sans-serif)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans']

# Bar plot for percentage of foreign nationals
plt.figure(figsize=(12, 6))
sns.barplot(x='Percentage', y='State', 
            data=foreign_percent.sort_values('Percentage', ascending=False))
plt.title('Percentage of Foreign Nationals by State in Germany (2022)')
plt.xlabel('Percentage (%)')
plt.ylabel('State')
plt.tight_layout()
plt.savefig('percentage_foreign_nationals.png', bbox_inches='tight')
plt.show()

# Bar plot for count of foreign nationals
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='State', 
            data=foreign_count.sort_values('Count', ascending=False))
plt.title('Number of Foreign Nationals by State in Germany (2022)')
plt.xlabel('Number of Foreign Nationals')
plt.ylabel('State')
plt.tight_layout()
plt.savefig('count_foreign_nationals.png', bbox_inches='tight')
plt.show()

# Observation
print("Observation: Bremen and Berlin have the highest percentage of foreign nationals, while Nordrhein-Westfalen and Bayern have the largest number of foreign nationals.")
```
