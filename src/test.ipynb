```python
import pandas as pd
import glob
import os

# مسیر پوشه‌ای که فایل‌های CSV داخلش هست
folder_path = "../data/raw"

# پیدا کردن همه فایل‌های CSV در پوشه
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

# لیست برای ذخیره دیتافریم‌ها
dfs = []
dfs = []

for file in all_files:
    try:
        df = pd.read_csv(file, sep=";")  # جداکننده ; چون داده‌هات اینجوری بود
        dfs.append(df)
    except Exception as e:
        print(f"مشکل در خواندن {file}: {e}")

# ترکیب همه دیتافریم‌ها
combined_df = pd.concat(dfs, ignore_index=True)

# ذخیره به عنوان یک فایل CSV جدید
output_path = os.path.join(folder_path, "combined_data.csv")
combined_df.to_csv(output_path, index=False)

print(f"فایل ترکیب‌شده ذخیره شد: {output_path}")
```

```python
import pandas as pd

# 1. فایل رو بخونیم
df = pd.read_csv("../data/raw/combined_data.csv")

# 2. فقط ستون‌های مهم رو ببینیم (مثلاً Destination = کشور مقصد، Count = تعداد مهاجران)
print(df.head())

# 3. داده‌ها رو بر اساس کشور مقصد گروه‌بندی کنیم
grouped = df.groupby("Destination")["Count"].sum().reset_index()

# 4. مرتب‌سازی نزولی (بیشترین مهاجران اول)
grouped = grouped.sort_values(by="Count", ascending=False)

# 5. جدول نهایی رو نشون بده
print(grouped)
```
